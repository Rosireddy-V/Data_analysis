Data analysis is a process of inspecting, cleaning, transforming and modeling data with the goal of discovering useful information, informing conclusions and supporting decision making

Data analysis process consists of the following steps:- 
1) Asking questions
2) Data wrangling
a) Gathering Data 
i) CSV Files
ii) API 
iii) Web Scraping 
iv) Databases 
b) Assessing Data 
c) Cleaning Data 
3) Exploratory Data analysis 
4) Drawing conclusions 
5) Communicating Results

Assessing Data:- 
1) finding the number of rows and columns (shape)
2) Data types of various columns (info())
3) Checking for missing values (info())
4) Check for duplicate data(is_unique)
5) Memory occupied by the dataset (info())
6) High level mathematical overview of the dataset (describe())

Data Quality Dimensions:- 

1) Completeness issues, ex:- No data
2) Validity issues, ex:- Invalid data like height in -ve values
3) Accuracy issues, ex:- Inaccurate values like height is 1cm
4) Consistency issues , ex:- Data is inconsistent like sometimes it is written newyork and sometimes as NY

Cleaning Data following order:- 
1) Completeness Data 
2) Tidiness issues (structural issues)
3) Remaining Data quality issues like validity, accuracy and consistency.

Cleaning Data:- 
1) Missing Data (mean)
2) Remove duplicate data (drop_duplicates)
3) Incorrect data type (astype)

Exploring Data:- 
1) finding correlation and covariance 
2) Doing univariate and multivariate analysis 
3) plotting graphs (data visualization)

Augumenting Data:- 
1) Removing outliers
2) Merging dataframes
3) Adding a new column
, These operations are collectively called as feature engineering

Exploratory Data Analysis

1) Analysis 
a) Univariate analysis
b) Multivariate analysis 


2) Feature Engineering 
a) Creating new columns 
b) Modifying existing ones

3) Handling Outliers 
a) Detect outliers 
b) remove outliers

```python
salary=[]
reviews=[]
rating=[]
interviews=[]
companyname=[]
about=[]
since=[]
ceo_names=[]
for i in range(1,200):
    url="https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}".format(i)
    tot_data_txt=requests.get(url,headers=headers).text
    tot_data=BeautifulSoup(tot_data_txt,'lxml')
    company=tot_data.find_all('div',class_='companyCardWrapper')
    for j in company:
        companyname.append(j.find('h2').text.strip())
        reviews.append(j.find_all('span',class_='companyCardWrapper__ActionCount')[0].text)
        salary.append(j.find_all('span',class_='companyCardWrapper__ActionCount')[1].text)
        interviews.append(j.find_all('span',class_='companyCardWrapper__ActionCount')[2].text)
        rating.append(j.find('span',class_='companyCardWrapper__companyRatingValue').text)
        try:
            about.append(j.find('span',class_='companyCardWrapper__interLinking').text.strip().split('|')[0])
        except:
            about.append(np.nan)
        k=j.find('meta',itemprop='url')
        new_url=k['content']
        r=requests.get(new_url,headers=headers).text
        overview=BeautifulSoup(r,'lxml')
        try:
            since.append(overview.find_all('p',class_='textItem__val aboutItem__value')[0].text.strip())
        except:
            since.append(np.nan)
        try:
            ceo_names.append(overview.find_all('p',class_='textItem__val aboutItem__value')[3].text.strip())
        except:
            ceo_names.append(np.nan)
```

```python
dic={'Company':companyname,'Reviews':reviews,'Salary':salary,'Rating':rating,'Interviews':interviews,'About':about,'Old':since,'CEO':ceo_names}
```

```python
data=pd.DataFrame(dic)
data.to_csv('mydata')
```
